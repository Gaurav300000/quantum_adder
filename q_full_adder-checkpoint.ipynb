```python
from qiskit import *
```

```python
# Creating Full Adder Circuit 
circuit = QuantumCircuit(4)
circuit.ccx(1,2,3)
circuit.cx(1,2)
circuit.ccx(0,2,3)
circuit.cx(0,2)
%matplotlib inline
circuit.draw(output = 'mpl')
```

```python
#Converting Above circuit into Quantum Gate
full_adder = circuit.to_gate()
#It maps |q0>|q1>|q2>|q3> -> |q0>|q1>|q0⊕q1>⊕q2>|(q1q2⊕q3)⊕q0(q1⊕q2)>
```

```python
c = QuantumRegister(4)
a = QuantumRegister(4)
b = QuantumRegister(5)
#Creating Register for storing four qubits state
#b has 5 qubit , one extra to store last i.e; fifth qubit of sum.
cr = ClassicalRegister(13)
```

```python
new_circuit = QuantumCircuit(c,a,b,cr)
#Creating new Quantum Circuit for adding two 4-qubit state
```

```python
for i in range(0,4):
    new_circuit.initialize([1,0],c[i])
new_circuit.initialize([1,0],b[4])    
#initializing carry qubits to 0 , and b4 = 0
```

```python
new_circuit.append(full_adder,[c[0],a[0],b[0],c[1]])
new_circuit.append(full_adder,[c[1],a[1],b[1],c[2]])
new_circuit.append(full_adder,[c[2],a[2],b[2],c[3]])
new_circuit.append(full_adder,[c[3],a[3],b[3],b[4]])
```

```python
new_circuit.measure(b[4],cr[12])
new_circuit.measure(b[3],cr[11])
new_circuit.measure(a[3],cr[10])
new_circuit.measure(c[3],cr[9])
new_circuit.measure(b[2],cr[8])
new_circuit.measure(a[2],cr[7])
new_circuit.measure(c[2],cr[6])
new_circuit.measure(b[1],cr[5])
new_circuit.measure(a[1],cr[4])
new_circuit.measure(c[1],cr[3])
new_circuit.measure(b[0],cr[2])
new_circuit.measure(a[0],cr[1])
new_circuit.measure(c[0],cr[0])
# Based on measurement the output will be in order |s4>|s3>|a3>|c3'>|s2>|a2>|c2'>|s1>|a1>|c1'>|s0>|a0>|c0'=c0>
```

```python
%matplotlib inline
new_circuit.draw(output='mpl')
```

```python
'''Above quantum circuit maps :
|b4>|b3>|b2>|b1>|b0>|a3>|a2>|a1>|a0>|c3>|c2>|c1>|c0>  -> |s4>|s3>|a3>|c3'>|s2>|a2>|c2'>|s1>|a1>|c1'>|s0>|a0>|c0'=c0>'''
# s[i] =a[i]⊕b[i]⊕c[i] for i( <= 3) and s4 = b4
```
